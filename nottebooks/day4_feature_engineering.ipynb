```python
# ==========================================
# Day3 升级版（推荐）
# 标签定义：未来20日最大回撤超过10% = crash
# ==========================================

import pandas as pd
import numpy as np

# ------------------------------------------
# 第1步：读取数据
# ------------------------------------------
df = pd.read_csv(r"C:\Users\1\Desktop\项目\stock-data\Prices.csv")

# 日期格式
df["Date"] = pd.to_datetime(df["Date"])

# ------------------------------------------
# 第2步：宽表 -> 长表
# ------------------------------------------
df_long = df.melt(
    id_vars="Date",
    var_name="symbol",
    value_name="close"
)

# 排序
df_long = df_long.sort_values(
    ["symbol", "Date"]
).reset_index(drop=True)

# ------------------------------------------
# 第3步：计算日收益率
# ------------------------------------------
df_long["return"] = df_long.groupby("symbol")["close"].pct_change()

# ------------------------------------------
# 第4步：未来20日最大回撤
# 定义：
# 从今天价格出发，未来20日最低价相对今天跌多少
# ------------------------------------------
def future_mdd(x):
    future_min = x.shift(-1).rolling(20).min()
    return future_min / x - 1

df_long["future_mdd_20"] = (
    df_long.groupby("symbol")["close"]
    .transform(future_mdd)
)

# ------------------------------------------
# 第5步：Crash标签
# 最大回撤超过10%
# ------------------------------------------
df_long["crash"] = np.where(
    df_long["future_mdd_20"] < -0.10,
    1,
    0
)

# ------------------------------------------
# 第6步：查看结果
# ------------------------------------------
print(df_long[[
    "Date",
    "symbol",
    "close",
    "return",
    "future_mdd_20",
    "crash"
]].head(20))

print("标签分布：")
print(df_long["crash"].value_counts())

# ------------------------------------------
# 第7步：各股票 crash 次数
# ------------------------------------------
print("各股票 crash 次数：")
print(
    df_long.groupby("symbol")["crash"]
    .sum()
    .sort_values(ascending=False)
)

# ------------------------------------------
# 第8步：保存文件
# ------------------------------------------
df_long.to_csv(
    r"C:\Users\1\Desktop\项目\stock-data\day3_dataset_mdd.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Day3 升级完成，已保存 day3_dataset_mdd.csv")
```

```python
# ==========================================
# Day4 特征工程（10个特征完整版）
# 输入：day3_dataset_mdd.csv
# 输出：day4_features.csv
# ==========================================

import pandas as pd
import numpy as np

# ------------------------------------------
# 第1步：读取 Day3 数据
# ------------------------------------------
df = pd.read_csv(
    r"C:\Users\1\Desktop\项目\stock-data\day3_dataset_mdd.csv"
)

# 日期格式
df["Date"] = pd.to_datetime(df["Date"])

# 排序
df = df.sort_values(
    ["symbol", "Date"]
).reset_index(drop=True)

print("原始数据维度：", df.shape)

# ------------------------------------------
# 第2步：确保收益率存在
# ------------------------------------------
if "return" not in df.columns:
    df["return"] = df.groupby("symbol")["close"].pct_change()

# ------------------------------------------
# 第3步：动量类特征（3个）
# ------------------------------------------
# 过去5日涨跌幅
df["ret_5d"] = df.groupby("symbol")["close"].pct_change(5)

# 过去20日涨跌幅
df["ret_20d"] = df.groupby("symbol")["close"].pct_change(20)

# 过去60日涨跌幅
df["ret_60d"] = df.groupby("symbol")["close"].pct_change(60)

# ------------------------------------------
# 第4步：波动率特征（2个）
# ------------------------------------------
# 10日波动率
df["vol_10d"] = (
    df.groupby("symbol")["return"]
    .rolling(10)
    .std()
    .reset_index(level=0, drop=True)
)

# 20日波动率
df["vol_20d"] = (
    df.groupby("symbol")["return"]
    .rolling(20)
    .std()
    .reset_index(level=0, drop=True)
)

# ------------------------------------------
# 第5步：均线特征（2个）
# ------------------------------------------
# 5日均线
df["ma_5"] = (
    df.groupby("symbol")["close"]
    .rolling(5)
    .mean()
    .reset_index(level=0, drop=True)
)

# 20日均线
df["ma_20"] = (
    df.groupby("symbol")["close"]
    .rolling(20)
    .mean()
    .reset_index(level=0, drop=True)
)

# 当前价格相对均线偏离
df["ma_gap_5"] = (df["close"] - df["ma_5"]) / df["ma_5"]
df["ma_gap_20"] = (df["close"] - df["ma_20"]) / df["ma_20"]

# ------------------------------------------
# 第6步：回撤特征（1个）
# ------------------------------------------
rolling_max_20 = (
    df.groupby("symbol")["close"]
    .rolling(20)
    .max()
    .reset_index(level=0, drop=True)
)

df["drawdown_20"] = (
    df["close"] - rolling_max_20
) / rolling_max_20

# ------------------------------------------
# 第7步：RSI(14) 特征（1个）
# ------------------------------------------
delta = df.groupby("symbol")["close"].diff()

gain = delta.clip(lower=0)
loss = -delta.clip(upper=0)

avg_gain = (
    gain.groupby(df["symbol"])
    .rolling(14)
    .mean()
    .reset_index(level=0, drop=True)
)

avg_loss = (
    loss.groupby(df["symbol"])
    .rolling(14)
    .mean()
    .reset_index(level=0, drop=True)
)

rs = avg_gain / avg_loss

df["rsi_14"] = 100 - (100 / (1 + rs))

# ------------------------------------------
# 第8步：Z-score（1个）
# ------------------------------------------
rolling_mean_20 = (
    df.groupby("symbol")["close"]
    .rolling(20)
    .mean()
    .reset_index(level=0, drop=True)
)

rolling_std_20 = (
    df.groupby("symbol")["close"]
    .rolling(20)
    .std()
    .reset_index(level=0, drop=True)
)

df["zscore_20"] = (
    df["close"] - rolling_mean_20
) / rolling_std_20

# ------------------------------------------
# 第9步：删除缺失值
# ------------------------------------------
df = df.dropna().reset_index(drop=True)

print("特征工程后维度：", df.shape)

# ------------------------------------------
# 第10步：查看结果
# ------------------------------------------
print(df.head())

print("\n字段名：")
print(df.columns.tolist())

print("\n标签分布：")
print(df["crash"].value_counts())

# ------------------------------------------
# 第11步：保存文件
# ------------------------------------------
df.to_csv(
    r"C:\Users\1\Desktop\项目\stock-data\day4_features.csv",
    index=False,
    encoding="utf-8-sig"
)

print("\nDay4 完成，已保存 day4_features.csv")
```
